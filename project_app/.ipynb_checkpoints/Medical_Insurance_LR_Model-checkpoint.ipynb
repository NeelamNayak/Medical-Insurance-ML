# 4. Feature Engineering

#### 1. Encoding

```python
df1 = pd.DataFrame(df.drop('charges',axis=1))
```

```python
df1 = pd.get_dummies(df1)
df1['charges'] = df['charges']
display(df1.sample(5))
```

#### 2. Outlier Imputation

```python
plt.figure(figsize=(16,3.5))
sns.boxplot(df1['age'])
```

```python
plt.figure(figsize=(15,3.5))
sns.boxplot(df1['children'])
```

```python
plt.figure(figsize=(16,3.5))
sns.boxplot(df1['bmi'])
```

```python
# Outliers find using IQR method
q1 = df1['bmi'].quantile(0.25)
q2 = df1['bmi'].quantile(0.50)
q3 = df1['bmi'].quantile(0.75)
median = df1['bmi'].median()

iqr = q3 - q1

upper_tail = q3 + 1.5 * iqr
lower_tail = q1 - 1.5 * iqr

print("Q1 :", q1)
print("Q2 :", q2)
print("Q3 :", q3)
print("Median :",median)

print("upper_tail :", upper_tail)
print("lower_tail :", lower_tail)
```

```python
df1[['bmi']].loc[df1['bmi'] > upper_tail] # uppertail outliers
```

```python
df1[['bmi']].loc[df1['bmi'] < lower_tail] # lowertail outliers
```

```python
df1.loc[df1['bmi'] > upper_tail, 'bmi'] = upper_tail # Outlier imputation using uppertail value
```

```python
df1.sample(10)
```

```python
plt.figure(figsize=(16,5))
sns.boxplot(df1['bmi']) # After Imputation cheking for outliers 
```

# 6. Model Building

#### 1. Splitting the data into Traing data and Tesing data

```python
x = df1.drop('charges', axis= 1)  # independent variables
y = df1['charges']  # dependent variables
```

```python
# Train Test Split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=12)
```

```python
x_train.sample(10)
```

```python
y_train.sample(10)
```

```python
x_test.sample(10)
```

```python
y_test.sample(10)
```

```python
# Create the instance of model
model = LinearRegression()
```

```python
# Training the model
model.fit(x_train,y_train)    # In background gradient decent algorithm runs
```

```python
model.coef_ # Return different values for slopes(m)
```

```python
model.intercept_ # Return value for y intercept(c)
```

```python
s2 = pd.Series(model.coef_, index=x.columns)
s2.sort_values().plot(kind = "barh")
```

# 7. Model Evaluation

#### 1. Evaluation on training data

```python
y_pred_train = model.predict(x_train)    # Predicted values for training data
y_pred_train[:10]
```

```python
# Evaluation matrics

mse = mean_squared_error(y_train, y_pred_train)
print("Mean Sqaured Error :",mse)

rmse = np.sqrt(mse)
print("Root Mean Sqaured Error :",rmse)

mae = mean_absolute_error(y_train, y_pred_train)
print("Mean Absolute Error :",mae)

r_squared_value = r2_score(y_train, y_pred_train)
print("R Squared Value :",r_squared_value)

adj_r2  = 1 - (((1 - r_squared_value) * (x_train.shape[0] - 1)) / (x_train.shape[0] - x_train.shape[1] - 1))
print("Adjusted R Squared Value :",adj_r2)
```

#### 2. Evaluation on testing data

```python
y_pred_test = model.predict(x_test)   # Predicted values for testing data
y_pred_test[:10]
```

```python
# Evaluation matrics

mse = mean_squared_error(y_test, y_pred_test)
print("Mean Sqaured Error :",mse)

rmse = np.sqrt(mse)
print("Root Mean Sqaured Error :",rmse)

mae = mean_absolute_error(y_test, y_pred_test)
print("Mean Absolute Error :",mae)

r_squared_value = r2_score(y_test, y_pred_test)
print("R Squared Value :",r_squared_value)

adj_r2  = 1 - (((1 - r_squared_value) * (x_test.shape[0] - 1)) / (x_test.shape[0] - x_test.shape[1] - 1))
print("Adjusted R Squared Value :",adj_r2)
```

#### 3. Normality of residuals

```python
residuals = y_train - y_pred_train   # Difference between actual and predicted value
residuals
```

```python
# Using distplot Visualization method
sns.distplot(residuals)
```

```python
# Using Shapiro method of Hypothesis testing
from scipy.stats import shapiro

_ , p_val = shapiro(residuals)

print("P_Value:",p_val)

if p_val >= 0.05:
    print("Null Hypothesis is Accepted")
    print("--> Data is Normally Distributed")
    
else:
    print("Null Hypothesis is Rejected >> Alternate Hypothesis is Accepted")
    print("Data is NOT Normally Distributed")
```

```python
# Using QQplot Visualization method
import statsmodels.api as sm
sm.qqplot(residuals, line = '45', fit=True)
```

#### 4. Homoscedasticity

```python
sns.scatterplot(x = y_train, y= residuals)
```
